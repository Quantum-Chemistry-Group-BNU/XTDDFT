```python
import numpy as np
import pandas as pd
file = './result/'
dataset = 'quest#4/b3lyp-st-10eV-noFock-noDA/'
```

```python
sTDA = pd.read_csv(file+dataset+'energysTDA.csv', sep='[,\s]+', header=None, engine='python').to_numpy()
BDF = pd.read_csv(file+dataset+'energyBDF.csv', sep='[,\s]+', header=None, engine='python').to_numpy()
exptl = pd.read_csv(file+dataset+'energyexptl.csv', sep='[,\s]+', header=None, engine='python').to_numpy()
sutda = np.expand_dims(sTDA[:,1], axis=1)
sxtda = np.expand_dims(sTDA[:,2], axis=1)
xtda = np.expand_dims(BDF[:,1], axis=1)
xtddft = np.expand_dims(BDF[:,2], axis=1)
utddft = np.expand_dims(BDF[:,3], axis=1)
utda = np.expand_dims(BDF[:,4], axis=1)
quest4x = np.expand_dims(exptl[:,1], axis=1)
questdb = np.expand_dims(exptl[:,2], axis=1)
```

```python
quest = np.concatenate((utda,sutda,xtda,sxtda,questdb),axis=1)
me = np.mean(quest[:,:-1] - np.expand_dims(quest[:,-1],axis=1), axis=0)
mae = np.mean(np.abs(quest[:,:-1] - np.expand_dims(quest[:,-1],axis=1)), axis=0)
# mse = np.mean(np.square(quest[:,:-1] - np.expand_dims(quest[:,-1],axis=1)), axis=0)
maxe = np.max(np.abs(quest[:,:-1]-np.expand_dims(quest[:,-1],axis=1)), axis=0)
print(f"UTDA ME: {me[0]:6.3f}    sUTDA ME: {me[1]:6.3f}    XTDA ME: {me[2]:6.3f}    sXTDA ME: {me[3]:6.3f}")
# print(f"sXTDA MSE: {mse[0]:6.4f}    XTDA MSE: {mse[1]:6.4f}    sUTDA MSE: {mse[2]:6.4f}    UTDA MSE: {mse[3]:6.4f}")
print(f"UTDA MAE: {mae[0]:6.3f}    sUTDA MAE: {mae[1]:6.3f}    XTDA MAE: {mae[2]:6.3f}    sXTDA MAE: {mae[3]:6.3f}")
print(f"UTDA MAXE: {maxe[0]:6.3f}    sUTDA MAXE: {maxe[1]:6.3f}    XTDA MAXE: {maxe[2]:6.3f}    sXTDA MAXE: {maxe[3]:6.3f}")
# print(f"XTDA MAE: {mae[1]:1.4f}")
# print(f"sUTDA MAE: {mae[2]:1.4f}")
# print(f"UTDA MAE: {mae[3]:1.4f}")
quest = np.concatenate((np.expand_dims(sTDA[:,0], axis=1), quest), axis=1)
quest = pd.DataFrame(quest)
quest.columns = ['mol.', 'U-TDA', 'sU-TDA', 'X-TDA', 'sX-TDA', 'TBE']
quest.to_csv(file+dataset+'Questdbavtz.csv', sep=',', header=True, index=False)
```
