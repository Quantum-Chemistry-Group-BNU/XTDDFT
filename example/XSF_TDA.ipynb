```python
from pyscf import gto, dft, scf, ao2mo, lib, tddft,fci
from pyscf.scf import stability
from pyscf.tools import cubegen
import numpy as np
from pyscf.dft import numint
import h5py
from XSF_TDA import XSF_TDA
```

```python
# Calculating excited energies with XSF-TDA
mol = gto.M(                 
    atom = f'''
    F 0 0 0
    H 0 0 1.0''',
    basis = '6-31G',
    charge = 0,
    spin = 2,  # from triplet state 
    verbose = 5,
    symmetry='C2v'
)

# In XSF-TDA, we should use ROKS
mf = dft.ROKS(mol)
mf.irrep_nelec = {'A1': (4,2), 'B1':(1,1), 'B2':(1,1)}
mf.xc = 'bhandhlyp'
mf.kernel()

```

```python
# in ALDA0 kernel
sf_tda = XSF_TDA(mf)
tmp_e, values = sf_tda.kernel(nstates=10)
print(tmp_e) #  in eV
sf_tda.analyse()
```

```python
# in multicollinear kernel
sf_tda = XSF_TDA(mf,method=1)
tmp_e, values = sf_tda.kernel(nstates=10)
print(tmp_e) #  in eV
sf_tda.analyse()
```
